### Sets
#### A set is a built-in data type in Python used to store multiple unique items in a single variable.

#### Sets are unordered → no indexing or slicing.
#### Sets are mutable → you can add/remove elements.
#### Sets are unique → duplicates are automatically removed.
#### Sets are defined using curly braces {} or the set() constructor

## Creating a set 

```python
fruits = { "apples", "bananas", "cherries", "dates", "apples"}

# set of integers

numbers = set ([ 1, 2, 3, 4, 5, 1, 2])

numbers
```

## Common set methods

```python
fruits 
```

```python
fruits.add("mangoes")
fruits
```

```python
# add multiple elements

fruits.update(["kiwis, grapes"])
fruits
```

```python
fruits.remove("apples")
fruits
```

```python
fruits.remove("pineapples")  # This will raise a KeyError since "pineapples" is not in the set
fruits
```

```python
fruits.discard("pineapples")  # This will not raise an error even if "pineapples" is not in the set
fruits
```

```python
fruits
```

```python
removed_fruit = fruits.pop()  # This will remove and return an arbitrary element from the set
removed_fruit
```

```python
fruits.clear() # This will remove all elements from the set
fruits
```
