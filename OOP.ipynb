### Creating a class

```python
class Dog:
	# we use the __init__ method to define initialize the object data
	# we use the self method to differentiate btwn the objects you are creating
	def __init__(self, name, breed):
		self.name = name
		self.breed = breed

	def bark(self):
		return f"{self.name} says woof!"
```

```python
dog_1 = Dog("Buddy", "Golden Retriever")
print(dog_1.name)
print(dog_1.bark())
```

```python
dog_2 = Dog("Husky",  "German Shepherd")
print(dog_2.name)
print(dog_2.bark())
```

```python
class Car:
    def __init__(self, make, model, year, color):
        self.make = make
        self.model = model
        self.year = year
        self.color = color
    
    def start_engine(self):
        print(f"The {self.year} {self.make} {self.model} is now running")
```

```python
car_1 = Car("BMW", "X1", 2020, "Red")
print(car_1.model)
print(car_1.start_engine())
```

### Accessing Object Attributes and Methods:

```python
my_car = Car("Toyota", "Corolla", 2020, "Red")
print(my_car.make)  
print(my_car.model)  
my_car.start_engine()
```

### Modifying Object Attributes

```python
my_car.color = "Black"
print(my_car.color)
```

### Class attributes vs Instance attributes 
#### Class attributes - are going to be shared across all objects and created inside your class and outside methods 
#### Instance attributes - are going to be unique to objects 

```python
class Circle:
    pi = 3.14  # class attribute

    def __init__(self, radius):
        self.radius = radius  # instance attribute

    def area(self):
        return self.pi * self.radius ** 2

circle_1 = Circle(7)
print(circle_1.radius)
print(circle_1.pi)
print(circle_1.area())

circle_2 = Circle(5)
print(circle_2.radius)
print(circle_2.area())


    
```

### Class attributes and methods
##### Class Attributes: Defined within the class but outside any methods. Shared among all instances of the class.
##### Instance Attributes: Defined inside methods using self. Unique to each instance.
##### Class Methods: Defined using the @classmethod decorator and take cls as the first parameter.
##### Static Methods: Defined using the @staticmethod decorator. Do not take self or cls as parameters.

```python
class MathOperations:
    pi = 3.14  # class attribute

    @classmethod
    def circle_area(cls, radius):
        return cls.pi * radius ** 2

    @staticmethod
    def add(a, b):
        return a + b

#create the objects from the class
obj1 = MathOperations()
obj2 = MathOperations()

#Using the class method
area1 = obj1.circle_area(5)
area1

# Using the static method
sum1 = obj2.add(10, 20)
sum1
```

### 1. Object Creation
##### Creating employee object

```python
class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position

    def work(self):
        print(f"{self.name} is working")

# creating an object
emp1 = Employee("Melody", "Data Analyst")
emp2 = Employee("Alice", "Software engineer")
```

### 2. Object Instanstiation and initialization

```python
print(emp2.name)  
print(emp2.work())
```

### 3. Object Usage

```python
# Adding a Method to the Employee Class

class Employee:

    def __init__(self, name, position):
        self.name = name
        self.position = position
        print(f"Employee {self.name} has been created.")

    def work(self):
        print(f"{self.name} is working as a {self.position}.")
# Creating and using an Employee object
emp1 = Employee("John Doe", "Software Developer")
emp1.work()  # Output: John Doe is working as a Software Developer.
```

### 4. Object Destruction

```python
class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position

    def __del__(self):
        print(f"Employee {self.name} is being deleted")

# creating an object
emp1 = Employee("Melody", "Data Analyst")
emp2 = Employee("Alice", "Software engineer")

del emp2
```
