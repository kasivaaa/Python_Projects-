### Using SQLite in Python

#### 1. Import the Library

```python
import sqlite3
```

#### 2. Create a New Database and connect it

```python
# create a connection to the database 
conn = sqlite3.connect("school_1.db")

conn.execute("PRAGMA foreign_keys =ON") #Enforces the foreign keys 
cursor = conn.cursor()
```

#### 3.Create Students table

```python
cursor.execute("""
               CREATE TABLE IF NOT EXISTS students (
               id INTEGER PRIMARY KEY AUTOINCREMENT,
               first_name TEXT NOT NULL,
               last_name TEXT NOT NULL,
               age INTEGER,
               grade TEXT,
               email TEXT UNIQUE
               )
               """)
conn.commit()
```

#### 3.1 Add a new column

```python
cursor.execute("ALTER TABLE students DROP COLUMN phone")
```

#### 4. Insert a Record into students

```python
cursor.execute("""INSERT INTO students (first_name, last_name, age, grade,email)
                VALUES (?,?,?,?,?)
                """, ("Cate","Micheal",17, "10th", "cate.micheal@gmail.com"))
conn.commit()
              
```

```python
#Instering Multiple commands 
students = [
    ("Alice", "Smith", 18, "10th", "alice.smith@example.com"),
    ("John", "Doe", 15, "10th", "john.doe@example.com"),
    ("Emily", "Mwangi", 17, "10th", "emily.mwangi@gmail.com"),
    ("Melody", "Bonareri", 16, "10th", "melody@gmail.com"),
    ("Joy", "Kilaha", 15, "10th", "joy@gmail.com"),
    ("Jude", "Wandera", 17, "11th", "jude.@gmail.com"),
    ("Alice", "Odhiambo", 15, "9th", "alice.odhiambo@gmail.com"),
    ("Brian", "Mutua", 16, "10th", "brian.mutua@gmail.com"),
    ("Cynthia", "Mwende", 17, "11th", "cynthia.mwende@gmail.com")
]
cursor.executemany(""" INSERT INTO students (first_name, last_name,age,grade,email)
                  VALUES(?,?,?,?,?)
                 """, students)
conn.commit()
```

```python
#View different records 
cursor.execute("SELECT *FROM students")
students = cursor.fetchall()

for student in students:
    print(student)
```
