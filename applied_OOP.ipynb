#### 2. __str__(self) - Human-Readable Description

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, Age: {self.age}"
    
person_1 = Person("Alice", 30)
str(person_1)
```

### Object Comparison and Arithmetic
#### 4. __eq__(self, other) - Equality check (==)

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
p1 = Point(2, 3)
p2 = Point(7, 3)
print(p1 == p2)
```

#### 5.__ne__(self, other)

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __ne__(self, other):
        return not (self.x == other.x and self.y == other.y)

p1 = Point(2, 3)
p2 = Point(4, 5)
print(p1 != p2)  # Output: True
```

#### Difference btwn __ne__(self, other) and __eq__(self, other) - Equality check (==) is functionality the first one is used to show whether they are equal and the second shows whether they are not equal

### 6.__lt__(self, other)

#### Shows less than comparison

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __lt__(self, other):
        return self.value < other.value

n1 = Number(5)
n2 = Number(10)
print(n1 < n2)  # Output: True
```

### 7.  __add__ (self, other)

#### shows addition

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3)  # Output: Vector(4, 6)
```

#### Object-Oriented Programming Principles


#### 1.Encapsulation
#### there us restricted access to some attributes 

#### A. Public Members
##### Accessible: Everywhere (Inside the class, outside the class)

```python
class Employee:
    def __init__(self, name, salary):
        #Public Attribute 
        self.name = name
        self.salary = salary
#calling the method
    def show_details(self):
        print(f"Name: {self.name}, Salary: {self.salary}")

emp1 = Employee("Annette", 60000)
print(emp1.name)
print(emp1.show_details())
```

#### B. Protected Members 
##### Accessible: Inside the class and its Subclasses (Child classes).
##### Syntax: Prefix with a single underscore _

```python
class Company:
   def __init__(self):
         #protected member
        self._project = "Python project"

# Child class
class Employee():
    def __init__(self, name):
        self.name = name
        self._project = "Python project"

        super().__init__()

    def show_project(self):
        # Accessing the protected member
        print(f"{self.name} is working on: {self._project}")

emp1 = Employee("Mark")
emp1.show_project()
```

### C.Private Members 
#### Accessible: ONLY inside the class where they are defined.
#### Syntax: Prefix with double underscores __.

```python
class Employee():
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary # private member


    def get_salary(self):
        # Accessing the private member
        print(f"Salary is: {self.__salary}")

emp1 = Employee("Mark", 10000)
emp1.get_salary()
```

### Comprehensive Example: The "Safe" Employee System

```python
class Employee:
    # define initilization method
    def __init__(self, name, id, salary, project):
        self.name = name
        self.id = id
        self._project = project # protected member
        self.__salary = salary #private member


    #Getter method(get the private member)
    def get_salary(self):
        return self.__salary

    #Setter method
    def set_salary(self, new_amount):
        if new_amount > 0:
            self.__salary = new_amount
            print(f"Salary updated")
        else:
            print("Error: Salary cannot be negative")

emp1 = Employee('James', 100, 20000, "Data Analysis")
emp1.get_salary()
emp1.set_salary(1000)
```

### Abstraction -Hiding details 

```python
from abc import ABC, abstractmethod

class Car(ABC):
    #define the abstract method
    @abstractmethod
    def mileage(self):
        pass

class Tesla(Car):
    def mileage(self):
        print("The mileage is 30kmph (Electric)")

class Suzuki(Car):
    def mileage(self):
        print("The mileage is 25kmph (Petrol)")


tesla_1 = Tesla()
suzuki_1 = Suzuki()

tesla_1.mileage()
suzuki_1.mileage()
```

### Example 2: The Geometric Shapes (Polygons)

```python
from abc import ABC, abstractmethod

# 1. The Abstract Base Class
class Polygon(ABC):
    
    @abstractmethod
    def sides(self):
        """Every polygon must define how many sides it has."""
        pass

# 2. Concrete Classes
class Triangle(Polygon):
    def sides(self):
        print("Triangle has 3 sides")

class Pentagon(Polygon):
    def sides(self):
        print("Pentagon has 5 sides")

class Hexagon(Polygon):
    def sides(self):
        print("Hexagon has 6 sides")

class Square(Polygon): # Note: Class names should be PascalCase
    def sides(self):
        print("Square has 4 sides")

# --- EXECUTION ---
# Creating objects
t = Triangle()
s = Square()
p = Pentagon()
k = Hexagon()

# Invoking the common interface
t.sides() # Output: Triangle has 3 sides
s.sides() # Output: Square has 4 sides
p.sides() # Output: Pentagon has 5 sides
k.sides() # Output: Hexagon has 6 sides
```

### 2. Method Overriding (Polymorphism with Inheritance)

```python
class Animal:
    def speak(self):
        return "Some sound"


class Dog(Animal):
   def speak(self):
        return "Bark!"
   
class Cat(Animal):
    def speak(self):
        return "Meow!"
    
dog = Dog()
cat = Cat()

print(dog.speak())
print(cat.speak())
```

```python
class Bird:
    def intro(self):
        print("There are many types of birds.")

    def flight(self):
        print("Most birds can fly but some cannot.")

class Sparrow(Bird):
    def flight(self):
        print("Sparrows can fly.")

class Ostrich(Bird):
    def flight(self):
        print("Ostriches cannot fly.")

# Polymorphic Function
def activate_bird(bird_obj):
    # This function works for ANY bird
    bird_obj.intro()
    bird_obj.flight()

# Usage
s = Sparrow()
o = Ostrich()

activate_bird(s) # Output: Sparrows can fly.
activate_bird(o) # Output: Ostriches cannot fly.
```
