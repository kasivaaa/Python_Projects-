```python
import re
```

### 1. re.search()

```python
text = "Python is fun. Python is powerful. Python is easy to learn. Many students love Python."
match = re.search(r"Python", text)
print(match)
```

```python
text = "Python is fun. Python is powerful. Python is easy to learn. Many students love Python."
match = re.search(r"Python", text)
print(match.group())
```

```python
print("This is my first line.\\n")#shows that n is printed as a string
```

```python
text = " My phone number is +254-071-123-54 and 0712856473"
regex = r'[+\d-]+'
match = re.search(regex, text)
match.group()
```

```python
text = "I was born on June 24"
pattern = r'([a-zA-Z]+) (\d+)'
match = re.search(pattern, text)
match.group()
```

### 2. re.findall()

```python
sentence = text = "Python is fun. Python is powerful. Python is easy to learn. Many students love Python."
```

```python
re.findall(r"Python", sentence)
```

```python
string = "Hello my numbers are 071654768 and 071587654"
pattern = r'\d+'
match = re.findall(pattern, string)
print(match)
```

```python
emails = "nicole@gmail.com, kali@gmail.com, kelvin@yahoo.com"
pattern = r'\w+@\w+\.'
match = re.findall(pattern, emails)
print (match)
```

### Extract dates

```python
text = " The event will be held on 25/12/2024 and 01/01/2025" 
pattern = r'\d{2}/\d{2}/\d{4}'
match = re.findall(pattern, text)
print(match)
```

```python
text = "Visit us at https://www.example.com or http://example.org for more information"
pattern = r'https?://[^\s]+'
match = re.findall(pattern, text)
print(match)
```

#### 4. re.split()

```python
data = "apple,banana;grape+orange"
fruits = re.split(r'[,;+]', data)
print(fruits)
```

#### 5. re.compile()

```python
pattern = re.compile ('[a-eA-E]')
print(pattern.findall("Aye, said Davido"))
```

```python
messy_data = """\
Customer Name: Jane Doe, Email: janedoe@gmail.com, Phone: +254-712-345678
Name: John Smith; Email: john_smith@company.co.ke; Tel: 0712345678
Customer=Alice; Email=alice@my-mail.org, Contact=+254712345679
Username: mike.1990! Email --> mike1990@yahoo.com Number: 0722 334 445
random text here, no useful data

Steve | steve@outlook.com | 0700-112-233 | Nairobi
"""
```

```python

# extract the emails
email_pattern = r'[\w.-]+@[\w.-]+\.\w+'
emails = re.findall(email_pattern, messy_data)

phone_number_pattern = r'(?:\+254[- ]?\d{3}[- ]?\d{6}\d{3}\)'
phones = re.findall(phone_number_pattern, messy_data)
print(emails)
print(phones)
```
