#### STRINGS IN PYTHON

##### What is a String?
##### A string is a data type in Python used to represent text. It's a sequence of characters enclosed in:

##### Single quotes ' '
##### Double quotes " "
##### Triple quotes ''' ''' or """ """ (for multi-line text)

```python
single = 'Hello World'
double = "World"
triple = "'this is a triple quoted string'"
```

### String slicing
#### String slicing allows you to extract a portion (substring) from a string using a range of indices.
#### Instead of accessing one character, slicing returns multiple characters.
#### Slicing Syntax:
#### string[start : stop : step]
#### start - index to begin (inclusive)
#### stop - index to end (exclusive)
#### step - interval between characters (optional)

```python
s = 'programming'
```

```python
s[0:4]
```

```python
# to get everything
s[:]
```

```python
s[4:]
```

### Reversing a string

```python
s[::-1]
```

### Looping through Strings
##### Looping through a string means accessing each character one by one and performing an action on it.

##### Since strings are iterable, you can loop through them easily in Python.

### Using a for Loop

```python
fruit = "banana"

for i in fruit:
    print(i)
```

### Looping through indexing

```python
#to get the index position
word = "python"
for i in range(len(word)):
    print(i)
```

```python
#to print out individual letters 
word = "python"
for i in range(len(word)):
    print(word[i])
```

### Using a while loop

```python
#looping through the word using a while loop
word = "python"
i=0
while i < len(word):
    print(word[i])
    i += 1
```

### Looping through a string with conditions

#### Print only using vowels

```python

    # use a for loop or a while loop to print out the vowels 

    words = "education"
    for char in words:
        if char.lower() in "aeiou":
            print(char)

```

### FILE HANDLING

### open ()- Opening a file

```python
file = open("example_1.text'", 'w')
```

### write() -Writing inside the file

```python
with open("example_1.txt", "w") as file:
	file.write("This is my first line.\n")
	file.write("This is my second line.\n")
```

### read() -Reading entire file content

```python
file = open('example_1.txt', 'r')
content = file.read()
print(content)
file.close()
```

### readline() - read an single line 

```python
file = open('example_1.txt', 'r')
content = file.readline()
print(content)
```

### readlines() - read an multiple lines

```python
file = open('example_1.txt', 'r')
content = file.readlines()
print(content)
```

### Using with – Automatically Handles close()
#### Best practice for working with files
#### Automatically calls file.close() for you

```python
with open('example_1.txt', 'r') as file:
    content = file.read()
    print(content)
```
