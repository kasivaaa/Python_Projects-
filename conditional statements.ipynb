```python
def calculate_grade(score):
    if score >= 90 and score <= 100:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    elif score >= 0:
        return "F"
    else:
        return "Invalid Score"

# Example usage
score = int(input("Enter your score: "))
print("Grade:", calculate_grade(score))
```

## Recurssive Function

```python
## the factorial 

def factorial (n):
    #the base case : If n is 1 
    if  n == 1:
        return 1
#Recursive case : n*(n-1)
    else:
        return n * factorial(n - 1)
    
print (factorial(4))
```

#### Example 2


```python
#fibonacci numbers 
def fibonacci(n):
    #base case : the first two positions (0,1 ) are just going to be the numbers themselves 
    if n <=1:
        return n
    # recursive case : the sum of the two preceding numbers
    else:
          return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(6))
  
```
