### 1. Import Libraries

```python
import requests
from bs4 import BeautifulSoup
```

### 2. Send HTTP Requests

```python
response = requests.get("https://books.toscrape.com/")
response
```

### 3. Parse HTML content

```python
# Parse HTML Content

soup = BeautifulSoup(response.content, 'html.parser')
soup
```

### Find book elements

```python
soup
```

```python
#find headings
header = soup.find("title")
header
```

```python
books = soup.find_all('article' , class_ ='product_pod')
books
```

### 5. Extract book data

```python


data = []
for book in books:
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text
    availability = book.find('p', class_='instock availability')
    data.append((title, price, availability))

data
```

## 6. Write data on a csv

```python
import csv
```

```python
with open('bookstore.csv', 'w')as file:
    writer = csv.writer(file)
    writer.writerow(['Title','Price','Availability'])
    writer.writerows(data)

    print("Data saved to bookstore.csv")
```

#### Use a while loop

```python
#Start explicity at page 2
url = "https://books.toscrape.com/catalogue/page-2.html"

while True:
    print(f"Scrappping: {url}")
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    #Extract data
    books = soup.select(".product_pod")
     
     #Moving to the next page 
    next_button = soup.select_one("li.next a")

    if next_button:
        next_page_url = next_button["href"]
        # construct the full url

        url = "https://books.toscrape.com/catalogue/"  + next_page_url
        time.sleep(2)

    else:
        print("No more pages found")
        break
```

### The solution : try/except blocks

```python
# Dummy HTML simulating a broken book card on page 3
html = """
<div class="product">
<h2>The Book of Python</h2>
<!-- Price is missing here! -->
</div>
"""



soup = BeautifulSoup(html, "html.parser")
card = soup.select_one(".product")

item = {}

try:
    item['title'] = card.select_one("h2").get_text(strip=True)
except AttributeError:
    item['title'] = "Unknown Title"

#Price
try:
    item['price'] = card.select_one(".price").get_text(strip=True)
except AttributeError:
    item['price'] = "N/A"
    print("Warning: The price is missing")

print(item)
```

### Network Reliability (Status codes )

```python
response = requests.get(url)

try:
    response.raise_for_status()
    print("Request successful")
except requests.exceptions.HTTPError as err:
    print(f"HTTP error occurred: {err}")
except requests.exceptions.RequestException as err:
    print(f"Request failed: {err}")
   
```
